## 04.元素記号

```python
string="Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
str_list=string.split(" ")
cl=[1,5,6,7,8,9,15,16,19]
dictionary={}

for k,strs in enumerate(str_list):
    if k+1 in cl:
        dictionary[strs[0:1]]=k+1
    else:
        dictionary[strs[0:2]]=k+1
    
print(dictionary)
        
```

## 05. n-gram

```python
def ngram (string,t):
    length= len(string)
    gram=[]
    for k in range(length-t+1):
        gram.append(string[k:t+k])
    return gram
        
nlper="I am an NLPer"
nlper_list = nlper.split(" ")

print("単語b-gram",ngram(nlper_list,2))
print("文字bi-gram",ngram(nlper,2))
```

## 06.集合

```python
str1="paraparaparadise"
str2 = "paragraph"
X_list = ngram(str1,2)
Y_list = ngram(str2,2)

X = set(X_list)
Y = set(Y_list)
print("X:",X)
print("Y:",Y)
print("和集合:",X|Y)
print("積集合",X&Y)
print("差集合",X-Y)

if 'se' in X:
    print("seはXに入ってる！！！")
else:
    print("seはXに入ってない！！！")
if 'se' in Y:
    print("seはYに入ってる！！！")
else:
    print("seはYに入ってない！！！")
    
```

## 08.暗号文

```python
def cipher(string):
    ret_str=""
    for str1 in string:
        k=ord(str1)
        if 97 <=  k<=122:
            ret_str+= str(k)
        else:
            ret_str+=str1
    return ret_str
        
a=cipher("RuuChanKawaii")
print(a)
```

## 09. Typoglycemia

```python
import random  

k="Banner Ruu AnkaK Kapibara Takuro Vier"
k_list = k.split(" ")
new_list = []
for strs in k_list :
    if len(strs) <= 4:
        new_list.append(strs)
    else:
        ret=list(strs[1:-1])
        random.shuffle(ret)
        newstr=strs[0]+''.join(ret)+strs[-1]
        new_list.append(newstr)
        
print(new_list)
```

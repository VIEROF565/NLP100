## 10.行数のカウント

## 11.タブをスペースに置換

```python
with open("popular-names.txt",'r') as f:
    for line in f:
        print(line.replace('\t',' '),end=' ')
```

## 12.１列目をcol1.txtに、2列目をcol2.txtに保存

```python
with open("popular-names.txt","r") as f, open("col1.txt","w") as f1, open("col2.txt","w") as f2:
        for line in f:
            words=line.split('\t')
            f1.write(words[0]+"\n")
            f2.write(words[1]+"\n")
            
```

## 13.col1.txtとcol2.txtをマージ

```python
with open("col1.txt","r") as f1, open("col2.txt","r") as f2, open("merge.txt","w") as mg:
    for line_f1,line_f2 in zip(f1,f2):
        arrays=[line_f1.replace("\n",""),line_f2.replace("\n","")]
        line_mg = "\t".join(arrays)
        mg.write(line_mg+"\n")
```

## 14.先頭からN行を出力

```python
N= int (input('N='))

with open("popular-names.txt","r") as f:
    for idx,line in enumerate(f):
        if idx < N:
            print(line)
        else:
            break
```

## 15.末尾のN行を出力

```python
N= int (input('N='))
lines=[]
with open("popular-names.txt","r") as f:
    for line in f:
        lines.append(line)
    target=lines[-N:]

for word in target:
    print(word)
```

## 16.ファイルをN分割する

```python
N = int(input('N='))
i=0
num=1
with open("popular-names.txt","r") as f:
    ch=open("child_1.txt","w")
    for line in f:
        if i ==N:
            ch.close()
            num+=1
            ch=open("child_{}.txt".format(num),"w")
            i=0
        ch.write(line)
        i+=1
```

## 17.1列目の文字列の異なり

```python
SET=set()

with open("col1.txt","r") as f:
    for line in f:
        SET= SET | {line.replace("\n","")}
        
print(SET)
```

## 18.各行を3コラム目の数値の降順にソート

```python
def choice3(string):
    arr=string.split("\t")
    return int(arr[2])

numarray=[]
with open("popular-names.txt","r") as f:
    s=f.readlines()
    
s.sort(key= choice3,reverse=True)
for line in s:
    print(line)
```

## 19.各行の1コラム目の文字列の出現頻度を求め、出現頻度の高い順に並べる

```python
arrays=[]
word2num={}

def getnum(string):
    words=string.split("\t")
    return word2num[words[0]]

with open("popular-names.txt") as f:
    for line in f:
        arrays.append(line) #各行をそのまま配列に格納
        arr=line.split("\t")
        if arr[0] in word2num:
            word2num[arr[0]]+=1
        else:
            word2num[arr[0]]=1
           
        
dict2=sorted(word2num.items(), key = lambda x:x[1],reverse=True)
for word in dict2:
    print(word[0],"({})".format(word[1]))
```
